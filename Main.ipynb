```python
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.data import loadlocal_mnist
from sklearn.neural_network import MLPClassifier
```

```python
def load_data():
    train_x, train_y= loadlocal_mnist('./dataset/train-images.idx3-ubyte','./dataset/train-labels.idx1-ubyte')
    test_x, test_y = loadlocal_mnist('./dataset/t10k-images.idx3-ubyte','./dataset/t10k-labels.idx1-ubyte')
    train_x = train_x / 255.
    test_x = test_x / 255.
    return train_x, train_y, test_x, test_y
```

```python
x_train, y_train, x_test, y_test= load_data()

print("x_train's shape: " + str(x_train.shape))
print("x_test's shape: " + str(x_test.shape))
print("y_train's shape: " + str(y_train.shape))
print("y_test's shape: " + str(y_test.shape))
```

```python
n_layers = 5
layer_sizes = [784, 256, 128, 64, 10]
learning_rate = 0.1
weight_init = 'normal'
batch_size = 200
num_epochs = 100
```

```python
relu_net = MyNeuralNetwork(n_layers, layer_sizes, 'relu',
                           learning_rate,weight_init,batch_size,num_epochs)
relu_net.set_test_switch(True)
relu_net.set_tsne_switch(True)
relu_net.assign_test_sets(x_test, y_test)
relu_net.fit(x_train, y_train)
```

```python
relu_net.plot_loss_graphs()
```

```python
relu_net.score(x_test, y_test)
```

```python
relu_net.plot_tsne(3)
```

```python
relu_net.save_model('relu')
```
